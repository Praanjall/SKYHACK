```python
# Let's start by loading the dataset and inspecting its contents to understand its structure
import pandas as pd

# Load the dataset
file_path = 'uaEDA.csv'
df = pd.read_csv(file_path)

# Display the first few rows and basic information about the dataset
df.head(), df.info()
```

```python
# Statistical summary of the numeric columns
numeric_summary = df.describe()

# Distribution of 'primary_call_reason' (categorical) and 'NumberOfCalls'
call_reason_distribution = df['primary_call_reason'].value_counts()
number_of_calls_summary = df.groupby('primary_call_reason')['NumberOfCalls'].sum()

# Let's visualize the relationship between 'primary_call_reason', 'total_time', and 'AHT'
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size for better readability
plt.figure(figsize=(12, 4))

# Plot the Number of Calls for each call reason
plt.subplot(1, 2, 1)
sns.barplot(x='NumberOfCalls', y='primary_call_reason', data=df, palette="Blues_d")
plt.title('Number of Calls by Call Reason')
plt.xlabel('Number of Calls')
plt.ylabel('Call Reason')

# Plot the Average Handling Time (AHT) for each call reason
plt.subplot(1, 2, 2)
sns.barplot(x='AHT', y='primary_call_reason', data=df, palette="Greens_d")
plt.title('AHT by Call Reason')
plt.xlabel('Average Handling Time (seconds)')
plt.ylabel('Call Reason')

plt.tight_layout()
plt.show()

# Display the summary statistics
numeric_summary, call_reason_distribution, number_of_calls_summary
```

```python
import pandas as pd

# Load the dataset to check its structure and perform exploratory data analysis (EDA)
file_path = 'uaeda1.csv'
data = pd.read_csv(file_path)

# Display the first few rows and basic info about the dataset
data_info = data.info()
data_head = data.head()

data_info, data_head
```

```python
# Checking for missing values
missing_values = data.isnull().sum()

# Descriptive statistics for numerical columns
descriptive_stats = data.describe()

missing_values, descriptive_stats
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Create subplots for visualizations
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Visualization 1: Number of Calls by primary_call_reason
sns.barplot(x='NumberOfCalls', y='primary_call_reason', data=data, ax=axes[0, 0], palette="viridis")
axes[0, 0].set_title('Number of Calls by Primary Call Reason')
axes[0, 0].set_xlabel('Number of Calls')
axes[0, 0].set_ylabel('Primary Call Reason')

# Visualization 2: Average Handle Time (AHT) by primary_call_reason
sns.barplot(x='AHT', y='primary_call_reason', data=data, ax=axes[0, 1], palette="magma")
axes[0, 1].set_title('Average Handle Time (AHT) by Primary Call Reason')
axes[0, 1].set_xlabel('Average Handle Time (AHT)')
axes[0, 1].set_ylabel('Primary Call Reason')

# Visualization 3: Call Duration by primary_call_reason
sns.barplot(x='Call_Duration', y='primary_call_reason', data=data, ax=axes[1, 0], palette="coolwarm")
axes[1, 0].set_title('Call Duration by Primary Call Reason')
axes[1, 0].set_xlabel('Call Duration')
axes[1, 0].set_ylabel('Primary Call Reason')

# Visualization 4: Total Time in Queue by primary_call_reason
sns.barplot(x='total_time_queue', y='primary_call_reason', data=data, ax=axes[1, 1], palette="Blues")
axes[1, 1].set_title('Total Time in Queue by Primary Call Reason')
axes[1, 1].set_xlabel('Total Time in Queue')
axes[1, 1].set_ylabel('Primary Call Reason')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Sorting the data by Call Duration and selecting the top 5 call reasons
top_10_call_reasons = data.sort_values(by='Call_Duration', ascending=False).head(10)

# Plotting the top 5 call reasons based on Call Duration
plt.figure(figsize=(10, 6))
sns.barplot(x='Call_Duration', y='primary_call_reason', data=top_10_call_reasons, palette="crest")
plt.title('Top 5 Call Reasons Based on Call Duration')
plt.xlabel('Call Duration')
plt.ylabel('Primary Call Reason')
plt.show()
```

```python
# Creating a line plot for the top 5 call reasons based on Call Duration
plt.figure(figsize=(10, 6))

# Sorting the data again to ensure the order is consistent for the line plot
top_5_call_reasons_sorted = top_10_call_reasons.sort_values(by='Call_Duration')

# Plotting the line plot
plt.plot(top_5_call_reasons_sorted['primary_call_reason'], 
         top_5_call_reasons_sorted['Call_Duration'], 
         marker='o', color='b', linestyle='-', linewidth=2)

plt.title('Top 5 Call Reasons Based on Call Duration')
plt.xlabel('Primary Call Reason')
plt.ylabel('Call Duration')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Let's first take a look at the contents of the uploaded file to understand its structure before performing EDA.
import pandas as pd

# Load the dataset
file_path = 'uaeda3.csv'
data = pd.read_csv(file_path)

# Display the first few rows and general info about the dataset
data_info = data.info()
data_head = data.head()

data_info, data_head
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Summary statistics
summary_stats = data.describe()

# Correlation matrix
corr_matrix = data.corr()

# Plotting the distributions of numerical columns
plt.figure(figsize=(14, 8))

# Distribution of NumberOfCalls
plt.subplot(2, 2, 1)
sns.histplot(data['NumberOfCalls'], kde=True, bins=20, color='blue')
plt.title('Distribution of NumberOfCalls')

# Distribution of sentiment_per_agent
plt.subplot(2, 2, 2)
sns.histplot(data['sentiment_per_agent'], kde=True, bins=20, color='green')
plt.title('Distribution of Sentiment per Agent')

# Distribution of silence_per_agent
plt.subplot(2, 2, 3)
sns.histplot(data['silence_per_agent'], kde=True, bins=20, color='red')
plt.title('Distribution of Silence per Agent')

# Correlation heatmap
plt.subplot(2, 2, 4)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')

plt.tight_layout()
plt.show()

summary_stats, corr_matrix
```

```python
# Dropping 'agent_id' to focus on the other features for further analysis
data_without_id = data.drop('agent_id', axis=1)

# Re-calculating correlation matrix and plotting heatmap without 'agent_id'
corr_matrix_without_id = data_without_id.corr()

# Re-plotting the correlation heatmap without 'agent_id'
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix_without_id, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation_Matrix')
plt.show()

corr_matrix_without_id
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare the data
X = data_without_id['NumberOfCalls'].values.reshape(-1, 1)
y_sentiment = data_without_id['sentiment_per_agent']
y_silence = data_without_id['silence_per_agent']

# Initialize linear regression models
model_sentiment = LinearRegression()
model_silence = LinearRegression()

# Fit the models
model_sentiment.fit(X, y_sentiment)
model_silence.fit(X, y_silence)

# Predicted values
y_sentiment_pred = model_sentiment.predict(X)
y_silence_pred = model_silence.predict(X)

# Plotting the results for Sentiment
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(X, y_sentiment, color='blue', label='Actual Sentiment')
plt.plot(X, y_sentiment_pred, color='red', label='Predicted Sentiment')
plt.title('Sentiment vs Number of Calls')
plt.xlabel('Number of Calls')
plt.ylabel('Sentiment')
plt.legend()

# Plotting the results for Silence
plt.subplot(1, 2, 2)
plt.scatter(X, y_silence, color='green', label='Actual Silence')
plt.plot(X, y_silence_pred, color='red', label='Predicted Silence')
plt.title('Silence vs Number of Calls')
plt.xlabel('Number of Calls')
plt.ylabel('Silence Percentage')
plt.legend()

plt.tight_layout()
plt.show()

# Displaying the model coefficients and intercepts
model_sentiment.coef_, model_sentiment.intercept_, model_silence.coef_, model_silence.intercept_
```

```python
# Let's load the CSV file and start by examining its structure to perform EDA.
import pandas as pd

# Load the dataset
file_path = 'ueda4.csv'
df = pd.read_csv(file_path)

# Display the first few rows and summary information of the dataset
df.head(), df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check for missing values
missing_values = df.isnull().sum()

# Scatter plot for Call_Duration vs sentiment_per_agent
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Call_Duration', y='sentiment_per_agent', data=df, color='blue')
plt.title('Scatter Plot of Call Duration vs Sentiment per Agent')
plt.xlabel('Call Duration (seconds)')
plt.ylabel('Sentiment per Agent')
plt.grid(True)
plt.show()

# Scatter plot for Call_Duration vs silence_per_agent
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Call_Duration', y='silence_per_agent', data=df, color='red')
plt.title('Scatter Plot of Call Duration vs Silence per Agent')
plt.xlabel('Call Duration (seconds)')
plt.ylabel('Silence per Agent')
plt.grid(True)
plt.show()

# Correlation matrix
correlation_matrix = df.corr()

missing_values, correlation_matrix
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare the data
X = df['Call_Duration'].values.reshape(-1, 1)

# Sentiment model
y_sentiment = df['sentiment_per_agent'].values
sentiment_model = LinearRegression()
sentiment_model.fit(X, y_sentiment)
predicted_sentiment = sentiment_model.predict(X)

# Silence model
y_silence = df['silence_per_agent'].values
silence_model = LinearRegression()
silence_model.fit(X, y_silence)
predicted_silence = silence_model.predict(X)

# Plotting the results
plt.figure(figsize=(16, 6))

# Sentiment vs Call Duration
plt.subplot(1, 2, 1)
plt.scatter(df['Call_Duration'], df['sentiment_per_agent'], color='blue', label='Actual Sentiment')
plt.plot(df['Call_Duration'], predicted_sentiment, color='red', label='Predicted Sentiment')
plt.title('Sentiment vs Call Duration')
plt.xlabel('Call Duration (seconds)')
plt.ylabel('Sentiment')
plt.legend()
plt.grid(True)

# Silence vs Call Duration
plt.subplot(1, 2, 2)
plt.scatter(df['Call_Duration'], df['silence_per_agent'], color='green', label='Actual Silence')
plt.plot(df['Call_Duration'], predicted_silence, color='red', label='Predicted Silence')
plt.title('Silence vs Call Duration')
plt.xlabel('Call Duration (seconds)')
plt.ylabel('Silence Percentage')
plt.legend()
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Let's first load and inspect the dataset to understand its structure, especially the 'primary call reason' and 'elite level code' columns.
import pandas as pd

# Load the dataset
file_path = 'ueda5.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
df.head()
```

```python
# Analyze the distribution of 'primary_call_reason'
primary_call_reason_counts = df['primary_call_reason'].value_counts()

# Analyze the distribution of 'elite_level_code'
elite_level_code_counts = df['elite_level_code'].value_counts(dropna=False)

# Display the top 10 call reasons and elite level code counts
primary_call_reason_counts.head(10), elite_level_code_counts.head(10)
```

```python
# Group by 'primary_call_reason' and 'elite_level_code' to analyze the relationship
call_reason_vs_elite = df.groupby(['primary_call_reason', 'elite_level_code']).size().unstack(fill_value=0)

# Display the relationship between the top 5 primary call reasons and elite level codes
call_reason_vs_elite.head(5)
```

```python
import matplotlib.pyplot as plt

# Plotting the distribution of top 5 primary call reasons with respect to elite level code
top_5_reasons = primary_call_reason_counts.head(5).index
filtered_data = call_reason_vs_elite.loc[top_5_reasons]

# Plot
filtered_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Primary Call Reasons vs Elite Level Code')
plt.xlabel('Primary Call Reason')
plt.ylabel('Number of Calls')
plt.legend(title='Elite Level Code')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting a vibrant color palette
colors = sns.color_palette("Set2", 6)

# Plotting the distribution of top 5 primary call reasons with respect to elite level code
top_5_reasons = primary_call_reason_counts.head(5).index
filtered_data = call_reason_vs_elite.loc[top_5_reasons]

# Create a stacked bar plot
filtered_data.plot(kind='bar', stacked=True, figsize=(10, 6), color=colors)
plt.title('Primary Call Reasons vs Elite Level Code', fontsize=16)
plt.xlabel('Primary Call Reason', fontsize=12)
plt.ylabel('Number of Calls', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Elite Level Code', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```
